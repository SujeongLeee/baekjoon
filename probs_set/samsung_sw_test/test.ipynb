```python
import math

# Constants
lambda_rate = 1.84  # jobs per hour
mu = 1  # jobs per hour per machine
m = 2  # number of machines
c_a2 = 1  # squared coefficient of variation of interarrival times
c_e2 = 0.25  # squared coefficient of variation of service times

# Traffic intensity
rho = lambda_rate / (m * mu)

# Pollaczek-Khintchine formula for M/G/m queue
E_R = (1 / mu) + ((c_a2 + c_e2) / 2) * (rho / (1 - rho)) * (1 / (m * mu))
``
# Convert hours to minutes for clarity
E_R_hours = E_R
E_R_minutes = E_R * 60

rho, E_R_hours, E_R_minutes
```
