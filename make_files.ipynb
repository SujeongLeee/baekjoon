```python
import os
import requests
from bs4 import BeautifulSoup

# Base URL for the steps
base_url = "https://www.acmicpc.net"

# URL of the '단계별로 풀어보기' page
steps_url = f"{base_url}/step"

# Add headers to the request
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

# Fetch the steps page
response = requests.get(steps_url, headers=headers)
response.raise_for_status()  # Ensure we notice bad responses

# Parse the steps page
soup = BeautifulSoup(response.text, "html.parser")



# Find all step links
step_rows = soup.select("table.table.table-bordered.table-striped tbody tr")


for step_row in step_rows:
    step_num = step_row.select_one("td:nth-child(1)").text.strip()

    step_link = str(step_row.select_one("td:nth-child(2)").select("a"))

    step_href = step_link.split("\"")[1]

    step_name = step_link.split(">")[1].split("<")[0].strip().replace(' ', '_')

    items_to_remove = [",", " ", ":"]
    for item in items_to_remove:
        step_name = step_name.replace(item, '')

    # Create directory name
    dir_name = f"{step_num}_{step_name}"
    os.makedirs(dir_name, exist_ok=True)
    print(f"Created directory: {dir_name}")



    # Fetch the step page
    step_url = f"{base_url}{step_href}"
    
    step_response = requests.get(step_url, headers=headers)
    step_response.raise_for_status()

    # Parse the step page
    step_soup = BeautifulSoup(step_response.text, "html.parser")

    # Find all problem rows
    problem_rows = step_soup.select("table.table.table-bordered.table-striped tbody tr")

    for problem_row in problem_rows:
        if len(problem_row) != 7:
            continue
        # Extract problem_step
        problem_step = problem_row.select_one("td:nth-child(1)").text.strip()

        # Extract problem number
        problem_number = problem_row.select_one("td.list_problem_id").text.strip()

        # Create .py file name
        file_name = f"{problem_step}_{problem_number}.py"
        file_path = os.path.join(dir_name, file_name)

        # Create an empty .py file
        with open(file_path, "w") as f:
            pass
        print(f"Created file: {file_path}")
```
